2.Find factorial of a number.

```
def factorial(n):
  if n == 0 or n == 1:
    return 1
  else:
    return n * factorial(n-1)

fact = factorial(5)
print(fact)
```

3.Print Fibonacci series.

```
n=10
a,b=0,1
for i in range(n):
  print(a,end=" ")
  a,b=b,a+b
```

4.Find largest among 3 numbers.

```
def max_val(a,b,c):
  return max(a,b,c)
print(max_val(3,6,9))
```
